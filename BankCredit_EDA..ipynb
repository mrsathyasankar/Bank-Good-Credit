# Exploratory Data Analysis

```python
#Import required libraries
import pandas as pd     
import numpy as np  
## visualisation
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#read data

data = pd.read_csv("data_for_eda.csv")
```

## Credit Cards Vs Bad Labels

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x=data.feature_1, hue=data.Bad_label)
plt.title("Credit Cards vs Bad Labels")
plt.xticks(rotation=90)
plt.show()
```

### Insights
* Most of the cusomers regardless of what kind of credit cards they hold, are having a good credit score
* A large number of people own Plainum Maxima, Platinum Deligh, and Titanium Deligh
* Veyr few people own the other credit cards
