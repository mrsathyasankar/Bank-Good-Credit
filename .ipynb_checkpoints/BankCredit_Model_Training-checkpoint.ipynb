# Project Overview: Predicting Credit Scores for GoodCredit Bank Customers

## Business Case

The goal of this data science project is to assist GoodCredit Bank in predicting credit scores for its current credit card customers. The credit score will serve as an indicator of a customer's creditworthiness, aiding the bank in mitigating credit default risk.

## Datasets Attributes

- **Target Variable:** Bad_label
  - 0: Customer has a good credit history
  - 1: Customer has a bad credit history (falls into 30 DPD + bucket)

The data comprises three tables: Cust_Account, Cust_Demographics, and Cust_Enquiry, each with its own set of columns.

### Tables Overview:

1. **Cust_Account:**
   - Historical accounts data and payment history.

2. **Cust_Demographics:**
   - Historical enquiry data, including enquiry amount and enquiry purpose.

3. **Cust_Enquiry:**
   - Current customer applications with demographic data.

Demographic features have been anonymized and renamed in compliance with privacy policies.

## Preface

In GoodCredit Bank, the evaluation of loan applications is based on merit, with a focus on determining the creditworthiness of individuals or entities. Assessing the level of risk associated with lending is crucial for making informed decisions. High-risk clients are less favorable due to a higher likelihood of default, while low-risk clients are more likely to have their loan applications approved.

## Domain Analysis

Credit analysis in banking involves scrutinizing the financial state, credit reports, and business cash flows of potential clients. The primary objective is to determine the default risk level a client presents and the potential losses the bank might incur in the event of default. The level of risk influences the approval or rejection of loan applications and, if approved, the amount to be awarded.

## Project Approach

The project involves the application of various machine learning models to predict credit scores. The models used include Logistic Regression, Decision Tree, Random Forest, K-Nearest Neighbors (KNN), Artificial Neural Network (ANN), Naive Bayes, and XGBoost.

## Data Cleaning and Exploration

During the data cleaning phase, necessary columns will be explained, and the datasets will be explored to understand the patterns and relationships within the data.

This project aims to provide GoodCredit Bank with a robust credit scoring system, allowing for more informed decision-making and reducing the risk of credit default.

# Converting Objects to Numbers using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder instance
label_encoder = LabelEncoder()

# Iterate through each column in the DataFrame
for column in data.columns:
    # Check if the column has object dtype (categorical)
    if data[column].dtype == 'object':
        # Fit and transform the LabelEncoder for each categorical column
        data[column] = label_encoder.fit_transform(data[column])

# Now, all categorical columns in 'data' are converted to numerical values
```

### Removing Highly correlated columns

```python

# Set the correlation threshold
correlation_threshold = 0.8

# Create the correlation matrix
correlation_matrix = data.corr()

# Find highly correlated pairs
highly_correlated_pairs = (abs(correlation_matrix) > correlation_threshold) & (correlation_matrix < 1)

# Create a set to store columns to drop
columns_to_drop = set()

# Iterate through each column in the DataFrame
for column in correlation_matrix.columns:
    # Check if the column is in the set of columns to drop
    if column not in columns_to_drop:
        # Find other columns highly correlated with the current column
        correlated_columns = list(correlation_matrix.loc[highly_correlated_pairs[column], column].index)
        
        # Add those columns to the set of columns to drop
        columns_to_drop.update(correlated_columns)

# Drop the highly correlated columns
data_dropped = data.drop(columns=columns_to_drop)

# Display the result
print(f"Original data shape: {data.shape}")
print(f"Data shape after dropping highly correlated columns: {data_dropped.shape}")
```

##  Train Test Split

```python
from sklearn.model_selection import train_test_split  # import train test split for spliting the data
# creating x_train,x_test,y_train,y_test
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=43) 
```

# We need to Balance Data as One value is dominating in y

```python
from imblearn.over_sampling import SMOTE  # assign SMOTE for handling unbalanced data
sm = SMOTE()
```

```python
x_sm,y_sm = sm.fit_resample(x_train,y_train)  # balancing data 
```

# Model Creation

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier   # assign decision tree classifier algorithmn
dt = DecisionTreeClassifier()                    
dt.fit(x_sm,y_sm)   # fit the model
```

```python
y_pred1=dt.predict(x_test) # predict the model
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
acc_1 = accuracy_score(y_pred1,y_test) # checking accuracy score
f1_1 = f1_score(y_pred1,y_test)
print("accuracy_score :", accuracy_score(y_pred1,y_test))
print("precision_score :", precision_score(y_pred1,y_test))
print("recall_score :", recall_score(y_pred1,y_test))
print("F1_score :", f1_score(y_pred1,y_test))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=100)
rf_clf.fit(x_sm,y_sm)
```

```python
y_pred2=rf_clf.predict(x_test)
```
