```python

# Convert 'opened_dt' column to datetime
data['opened_dt'] = pd.to_datetime(data['opened_dt'])

# Create new columns with year and month
data['opened_dt_year'] = data['opened_dt'].dt.year

# Drop the original 'opened_dt' column
data.drop('opened_dt', axis=1, inplace=True)
```

After this operation, interestingly there are 426 null values in opened_dt_year. We can replace it with mode

```python
# Calculate the mode value of 'opened_dt_year'
mode_opened_dt_year = data['opened_dt_year'].mode()[0]

# Fill missing values with the mode
data['opened_dt_year'].fillna(mode_opened_dt_year, inplace=True)
```

```python
# Convert 'feature_3' to numeric
data['feature_3'] = pd.to_numeric(data['feature_3'], errors='coerce')

# Replace values less than 500 with the median
data.loc[data["feature_3"] < 500, "feature_3"] = np.median(data['feature_3'].dropna())
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder instance
label_encoder = LabelEncoder()

# Iterate through each column in the DataFrame
for column in data.columns:
    # Check if the column has object dtype (categorical)
    if data[column].dtype == 'object':
        # Fit and transform the LabelEncoder for each categorical column
        data[column] = label_encoder.fit_transform(data[column])

# Now, all categorical columns in 'data' are converted to numerical values
```

```python
plt.figure(figsize=(28,28))#increase plot size
sns.heatmap(data.corr(),annot=True,cmap="RdYlGn")#its show corelation between each and every columns by plot heatmap
#here we can see that every columns are connected to each other,So we can't remove any columns.
```

```python

# Set the correlation threshold
correlation_threshold = 0.8

# Create the correlation matrix
correlation_matrix = data.corr()

# Find highly correlated pairs
highly_correlated_pairs = (abs(correlation_matrix) > correlation_threshold) & (correlation_matrix < 1)

# Create a set to store columns to drop
columns_to_drop = set()

# Iterate through each column in the DataFrame
for column in correlation_matrix.columns:
    # Check if the column is in the set of columns to drop
    if column not in columns_to_drop:
        # Find other columns highly correlated with the current column
        correlated_columns = list(correlation_matrix.loc[highly_correlated_pairs[column], column].index)
        
        # Add those columns to the set of columns to drop
        columns_to_drop.update(correlated_columns)

# Drop the highly correlated columns
data_dropped = data.drop(columns=columns_to_drop)

# Display the result
print(f"Original data shape: {data.shape}")
print(f"Data shape after dropping highly correlated columns: {data_dropped.shape}")
```
